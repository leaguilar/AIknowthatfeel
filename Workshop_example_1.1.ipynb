```python
import pandas as pd
import collections
import operator
import pickle
```

```python
def loadAndPrepareData(fname,vocab_size=5000,max_conv_length=50):
    data = pd.read_csv(fname,sep='	',header=None)
    x_data_raw=data.filter([1,2,3], axis=1).copy(deep=True)
    return x_data_raw
```

```python
def FlattenText(data):
    longText=''
    for index, row in data.iterrows():
        longText=longText+row.iloc[0]+' ### '+row.iloc[1]+' ### '+row.iloc[2]
    return longText
```

```python
x_train_raw=loadAndPrepareData('data_train/train.txt')
x_dev_raw=loadAndPrepareData('data_dev/dev.txt')
```

```python
len(x_train_raw)+len(x_dev_raw)
```

```python
x_concat=pd.concat([x_train_raw, x_dev_raw], ignore_index=True)
```

```python
len(x_concat)
```

```python
longText=FlattenText(x_concat)
```

```python
wordlist = longText.split()
```

```python
freq=collections.Counter(wordlist)
```

```python
dfreq=dict(freq)
```

```python
dfreq
```

```python
sorted_freq = reversed(sorted(dfreq.items(), key=operator.itemgetter(1)))
```

```python
sorted_freq
```

```python
i=0
embDict={}
for k,val in sorted_freq:
    embDict[k]=i
    i=i+1
```

```python
with open('models/wordDictionary.pkl', 'wb') as handle:
    pickle.dump(embDict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
